```python
import pandas as pd
df = pd.read_csv('C:/Users/User/Desktop/vllgyak/inputs (1).csv')
df.drop(['861_Data', '862_Data', '869_Data', '870_Data', '871_Data', '872_Data', '873_Data', '893_Data', '894_Data'], axis=1, inplace=True)

df.info()
df.plot()
```

```python
from statsmodels.tsa.stattools import adfuller
result = adfuller(df['856_Data'])
print('ADF Statistic:', result[0])
print('p-value:', result[1])
if result[1] > 0.05:
    print('The data is not stationary')
else:
    print('The data is stationary')
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import matplotlib.pyplot as plt
number_of_lags = 50
plot_acf(df['856_Data'], lags=number_of_lags)
plot_pacf(df['856_Data'], lags=number_of_lags)
plt.show()
```

```python
from statsmodels.tsa.arima.model import ARIMA
model = ARIMA(df['856_Data'], order=(1, 0, 1))
model_fit = model.fit()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
train_predictions = model_fit.predict(typ='levels')
train_mae = mean_absolute_error(df['856_Data'], train_predictions)
train_rmse = mean_squared_error(df['856_Data'], train_predictions, squared=False)

print('Training MAE:', train_mae)
print('Training RMSE:', train_rmse)
```

```python
future_steps = 5000
#alpha = 0.05
future_predictions = model_fit.forecast(steps=future_steps)
```

```python
print(future_predictions)
```

```python
# Plot the actual data
plt.plot(df['856_Data'], label='Actual')

# Plot the predicted values
plt.plot(train_predictions, label='Train Predictions')
plt.plot(future_predictions, label='Future Predictions')

plt.title('ARIMA Predictions')
plt.legend()

# Display the plot
plt.show()
```
